###shapiro wilk e qqplot

📌 Interpretazione del p-value:

Se p < 0.05, la distribuzione non è normale.
Se p ≥ 0.05, la distribuzione può essere considerata normale.

```python
import scipy.stats as stats

# Lista delle variabili numeriche
variabili_numeriche = ['elixsum3', 'eta', 'los', 'num_words']

# Eseguire il test di Shapiro-Wilk per ogni variabile numerica
for var in variabili_numeriche:
    stat, p = stats.shapiro(df[var])
    print(f"Test di Shapiro-Wilk per {var}: Statistica = {stat:.4f}, p-value = {p:.4f}")
```

📌 Interpretazione del Q-Q plot:

Se i punti seguono la diagonale, la distribuzione è normale.
Se ci sono deviazioni dalla diagonale, la distribuzione potrebbe essere asimmetrica o avere code pesanti.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm

fig, axes = plt.subplots(2, 2, figsize=(10,10))

for i, var in enumerate(variabili_numeriche):
    row, col = i // 2, i % 2
    sm.qqplot(df[var], line='s', fit=True, ax=axes[row, col])
    axes[row, col].set_title(f"Q-Q Plot di {var}")

plt.tight_layout()
plt.show()
```

## anova o kruskas wallis

Se p-value < 0.05, ci sono differenze significative tra i gruppi.

```python
import scipy.stats as stats
import pandas as pd

# Definizione delle variabili
variabili_numeriche = ['elixsum3', 'num_words', 'los', 'eta']
variabili_categoriche = ['cod_reg', 'anno_rif', 'disc']

# Ciclo per eseguire il test di Kruskal-Wallis
for cat in variabili_categoriche:
    print(f"\n--- Test di Kruskal-Wallis per la variabile categorica: {cat} ---")

    for num in variabili_numeriche:
        gruppi = [df[num][df[cat] == livello].dropna() for livello in df[cat].unique()]

        # Esegui il test solo se ci sono almeno due gruppi distinti
        if len(gruppi) > 1:
            stat, p = stats.kruskal(*gruppi)
            print(f"{num} vs {cat}: Statistica = {stat:.4f}, p-value = {p:.4f}")
        else:
            print(f"{num} vs {cat}: Non ci sono abbastanza gruppi per il test.")
```

```python
import pandas as pd
import scipy.stats as stats
from itertools import combinations

# Lista delle variabili categoriche
variabili_categoriche = ['cod_reg', 'cod_ist', 'reg_res', 'prov_res', 'com_res', 'cittad',
                         'mod_dim', 'disc', 'dpr_x', 'dsec1', 'dsec2', 'dsec3', 'dsec4',
                         'dsec5',   'female', 'anno_rif', 'target' ,'ynel1', 'ynel2',
       'ynel3', 'ynel4', 'ynel5', 'ynel6', 'ynel7', 'ynel8', 'ynel9', 'ynel10',
       'ynel11', 'ynel12', 'ynel13', 'ynel14', 'ynel15', 'ynel16', 'ynel17',
       'ynel18', 'ynel19', 'ynel20', 'ynel21', 'ynel22', 'ynel23', 'ynel24',
       'ynel25', 'ynel26', 'ynel27', 'ynel28', 'ynel29', 'ynel30', 'ynel31']

# Ciclo per confrontare tutte le coppie di variabili categoriche
for var1, var2 in combinations(variabili_categoriche, 2):
    # Crea la tabella di contingenza
    tabella_contingenza = pd.crosstab(df[var1], df[var2])

    # Controlla che non ci siano troppe celle vuote
    if (tabella_contingenza.shape[0] > 1) and (tabella_contingenza.shape[1] > 1):
        chi2, p, dof, expected = stats.chi2_contingency(tabella_contingenza)
        print(f"{var1} vs {var2}: Chi-quadro = {chi2:.4f}, p-value = {p:.4f}")
    else:
        print(f"{var1} vs {var2}: Non ci sono abbastanza dati per eseguire il test.")
```
